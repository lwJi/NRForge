## Auxiliary Variables

$$
\begin{align}
k_a
&=
\left(
    1, \frac{rx+ay}{r^2+a^2}, \frac{ry-ax}{r^2+a^2}, \frac{z}{r}
\right)
\\
f
&=\frac{r^2}{r^4+a^2z^2}2Mr
\\
g_{ab}
&=\eta_{ab}+fk_ak_b
\\
\beta_i
&=g_{0i}=fk_0k_i
\\
\alpha
&=
\left(
    1+\frac{2Mr^3}{r^4+a^2z^2}
\right)^{-1/2}
\end{align}
$$
